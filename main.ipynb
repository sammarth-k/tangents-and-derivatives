# Derivatives and Tangents

```python
# derivative of a function
def derivative(f,x):
    h = 0.000000000001
    d = (f(x+h) - f(x))/h
    
    return d
```

We know that the slope of the tangent of a function is equal to the value of the derivative at that point.
$$m = \frac{dy}{dx}$$

We use this value to get the equation of the tangent to the point via the point-slope form
$$y_t - y_0 = \frac{dy}{dx}(x_t - x_0)$$
$$y_t = \frac{dy}{dx}(x_t - x_0) + y_0$$

```python
# tangent of the point on a function
def tangent(f, x, x_0):
    y_0 = f(x_0)
    
    return derivative(f,x_0) * (x - x_0) + y_0
```

We can now start plotting

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches

# defining a function to plot the function and the derivative with the tangent at 
def plot_derivative(f,lowerx=0, upperx=10, lowery=-10, uppery=10,n=10):
    
    # values for the function
    x = np.linspace(lowerx, upperx,200)
    y = f(x) 

    # plotting the function
    plt.figure(figsize=(15,9))
    plt.plot(x,y,color="red")
    slopes=[]
    # plotting the tangents and points
    for i in np.linspace(lowerx, upperx, n):
        
        # plot tangent
        plt.plot(x,tangent(f, x, i),'b-', alpha = 0.1)
    
        plt.scatter(i, derivative(f,i), color="pink")
        
        
    plt.xlabel('x')     
    plt.ylabel('y')    
    
    plt.title('Plot of function, tangents at each point, derivative') 
    plt.xlim(lowerx,upperx)  
    plt.ylim(lowery,uppery)
    
    colors = ["r", "pink", "b"]
    texts = ["Function", "Derivative", "Tangent"]
    patches = [ mpatches.Patch(color=colors[i], label="{:s}".format(texts[i]) ) for i in range(len(texts)) ]
    plt.legend(handles=patches, ncol=2 )

    
    plt.show()
    plt.close()
```

```python
# define the function
def f(x):
    return np.sin(x)

# plot
plot_derivative(f,n=100)
```

# Interactive Mode

```python
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches

# define the function to use
def f(x):
    return np.sin(x)

# defining a function to plot the function and the derivative with the tangent at 
def plotter(i, lowerx=0, upperx=10, lowery=0,uppery=10):
    
    # values for the function
    x = np.linspace(lowerx, upperx,200)
    y = f(x) 
    
    plt.figure(figsize=(15,9))
    plt.plot(x,y,'r-')
    plt.plot(x, [derivative(f,i) for i in x], linestyle='dashed', color="black", alpha=0.4)
    
    plt.plot(x,tangent(f, x, i),'b-', alpha = 0.6)
    
    plt.scatter(i, derivative(f,i), color="black", alpha=1) 
        
    plt.xlabel('x')     
    plt.ylabel('y')    
    
    plt.title('Plot of function, tangents at each point, derivative') 
    plt.xlim(lowerx,upperx)  
    plt.ylim(lowery,uppery)
    
    
    colors = ["r", "black", "b"]
    texts = ["Function", "Derivative", "Tangent"]
    patches = [ mpatches.Patch(color=colors[i], label="{:s}".format(texts[i]) ) for i in range(len(texts)) ]
    plt.legend(handles=patches, ncol=2 )

    plt.show()
    plt.close()
    

@interact(i=np.pi,lowery=-100)
def plot(i):
    plotter(i, lowery=-2,uppery=2)
```
